```python
import pandas as pd
import matplotlib as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('RunBarChart_Tratado.xlsx')
```

```python
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df = df.sort_values(by=['Município', 'Data'])
```

```python
df['Quantidade_Cumulativa'] = df.groupby('Município')['Quantidade'].cumsum()
```

```python
df
```

```python
df = df.pivot_table('Quantidade_Cumulativa','Data', 'Município')
```

```python
df
```

```python
colunas = [
    'Abaeté', 'Betim', 'Brumadinho', 'Caetanópolis', 'Curvelo',
       'Esmeraldas', 'Felixlândia', 'Florestal', 'Fortuna de Minas', 'Igarapé',
       'Juatuba', 'Maravilhas', 'Morada Nova de Minas', 'Mário Campos',
       'Outros', 'Paineiras', 'Papagaios', 'Paraopeba', 'Pará de Minas',
       'Pequi', 'Pompéu', 'São Gonçalo do Abaeté', 'São Joaquim de Bicas',
       'São José da Varginha', 'Três Marias'
]
df[colunas] = df[colunas].fillna(method='ffill')
```
